**Solution:**

The following two equations are given in the question:

$\frac{\partial\mu}{\partial t} = \frac{\partial^2\mu}{\partial x^2}+a\frac{\partial\mu}{\partial x}+b\mu$ -(1)

$\mu(x,t) = e^{\alpha x+\beta t}v(x,t)$   -(2) 

For simplicity, I'm going to write (2) as $\mu = e^{\alpha x+\beta t}v$

From equation (1), we know we need to get $\frac{\partial\mu}{\partial t}$,$\frac{\partial^2\mu}{\partial x^2}$, and $\frac{\partial\mu}{\partial x}$

By using product and chain rule:

$\frac{\partial\mu}{\partial t} = \beta e^{\alpha x+\beta t}v+ e^{\alpha x+\beta t}\frac{\partial v}{\partial t}$  -(3)

$\frac{\partial\mu}{\partial x} = \alpha e^{\alpha x+\beta t}v+ e^{\alpha x+\beta t}\frac{\partial v}{\partial x}$  -(4)

$\frac{\partial^2\mu}{\partial x^2} = \alpha^2e^{\alpha x+\beta t}v+ \alpha e^{\alpha x+\beta t}\frac{\partial v}{\partial x}$
$+\alpha e^{\alpha x + \beta t}\frac{\partial v}{\partial x}+ e^{\alpha x + \beta t}\frac{\partial^2v}{\partial x^2}$ -(5)

Plug (3), (4), and (5) into (1):

$\beta e^{\alpha x+\beta t}v+ e^{\alpha x+\beta t}\frac{\partial v}{\partial t}$
$=\alpha^2e^{\alpha x+\beta t}v+ \alpha e^{\alpha x+\beta t}\frac{\partial v}{\partial x}$
$+\alpha e^{\alpha x + \beta t}\frac{\partial v}{\partial x}+ e^{\alpha x + \beta t}\frac{\partial^2v}{\partial x^2}$
$+a(\alpha e^{\alpha x+\beta t}v+ e^{\alpha x+\beta t}\frac{\partial v}{\partial x})$
$+ b(e^{\alpha x+\beta t}v)$


Rearrange:

$e^{\alpha x+\beta t}\frac{\partial v}{\partial t}$
$=(\alpha^2+a\alpha+b-\beta)e^{\alpha x+\beta t}v$
$+(2\alpha+a) e^{\alpha x+\beta t}\frac{\partial v}{\partial x}$
$+e^{\alpha x + \beta t}\frac{\partial^2v}{\partial x^2}$

Eliminate $e^{\alpha x+\beta t}$ on both sides:

$\frac{\partial v}{\partial t}$
$=(\alpha^2+a\alpha+b-\beta)v$
$+(2\alpha+a)\frac{\partial v}{\partial x}$
$+\frac{\partial^2v}{\partial x^2}$

Now, the goal is to reduce the above to $\frac{\partial v}{\partial t}=\frac{\partial^2v}{\partial x^2}$ 

That is, you have to somehow make $(\alpha^2+a\alpha+b-\beta)v$ and $(2\alpha+a)\frac{\partial v}{\partial x}$ disappear.

The only way to do that is to make:

$(\alpha^2+a\alpha+b-\beta)=0$    -(6)

and

$(2\alpha+a)=0$     -(7)

From (7), we get $\alpha=-\frac{1}{2}a$. Put this to (6):

$(-\frac{1}{2}a)^2-a\frac{1}{2}a+b-\beta =0$

So $\beta = b-\frac{1}{4}a^2$
